```python
from textblob import TextBlob

def analyze_sentiment(text):
    blob = TextBlob(text)  
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

def main():
    text = input("Enter a text string to analyze sentiment: ")
    sentiment = analyze_sentiment(text)
    print(f"The sentiment of the text is: {sentiment}")

if __name__ == "__main__":
    main()
```

```python
# Now we will test it on a Dataset: https://www.kaggle.com/datasets/abhi8923shriv/sentiment-analysis-dataset
import pandas as pd
from textblob import TextBlob

test_data = pd.read_csv('test.csv',encoding='latin1');

def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Test examples from the dataset
def test_sentiments():
    for index, row in test_data.iterrows():
        text = row['text']
        true_label = row['sentiment']
        predicted_sentiment = analyze_sentiment(text)
        print(f"Text: {text}")
        print(f"True Sentiment: {true_label}")
        print(f"Predicted Sentiment: {predicted_sentiment}")
        print()

if __name__ == "__main__":
    test_sentiments()
```
